```python
#Hottest month, coldest month against average
Temperatures = [29.6, 16.2]
Averages = [26.943333, 17.903333]
index = ['Warmest Month - January 2017', 'Coldest Month - July 1998']
df = pd.DataFrame({'Month Temperature (C)':Temperatures, 
'Month Average (C)': Averages}, index=index)
ax = df.plot.bar(rot=0, color={"Month Temperature (C)":"red","Month Average (C)":"darkorange"},title="Hottest month and coldest month against avg")
```

```python
#Temp and birth rates change over time (1990-2019)

y1 = Temp_data.mean(axis=1)
y2 = Birth_data.mean(axis=1)

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(y1, 'r-')
ax2.plot(y2, 'g-')

plt.title('Temperature and birth rates in NSW between 1990-2019')
ax1.set_xlabel('Year')
ax1.set_ylabel('Temperature (C)', color='r')
ax2.set_ylabel('Birth Rates', color='g')

plt.show()
#Increase in Birth rates as decrease in Temperature 
```

```python
#Yearly Birth Rates against Rainfall 

y1 = Rainfall_data.mean(axis=1)
y2 = Birth_data.mean(axis=1)

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(y1, 'b-')
ax2.plot(y2, 'g-')

plt.title('Rainfall and birth rates in NSW between 1990-2019')
ax1.set_xlabel('Year')
ax1.set_ylabel('Rainfall (mm)', color='b')
ax2.set_ylabel('Birth Rates', color='g')

plt.show()
#Decrease in birth rates as increase in Rainfall 
```

```python
y1 = Temp_data.mean(axis=1)
y2 = co2_data

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(y1, 'r-')
ax2.plot(y2, 'y-')

plt.title('C02 levels in Australia and temperature in NSW between 1990-2019')
ax1.set_xlabel('Year')
ax1.set_ylabel('Temperature (C)', color='r')
ax2.set_ylabel('CO2(ppm)', color='y')

plt.show()
#Increase in Birth rates as decrease in Temperature 
```

```python
#yearly fertility rates in aus vs temp in aus
y1 = Temp_data.mean(axis=1)
y2 = fert_data

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(y1, 'r-')
ax2.plot(y2, 'purple')

plt.title('Temperature in NSW and Fertility rates in Australia between 1990-2019')
ax1.set_xlabel('Year')
ax1.set_ylabel('Temperature (C)', color='r')
ax2.set_ylabel('Fertility Rate', color='purple')

plt.show()
# 
```
