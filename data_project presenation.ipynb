```python
#Hottest month, coldest month against average
Temperatures = [29.6, 0.0]
Averages = [26.943333, 17.903333]
index = ['Warmest Month - January 2017', 'Coldest Month - July 1998']
df = pd.DataFrame({'Month Temperature (°C)':Temperatures, 
'Month Average (°C)': Averages}, index=index)
ax = df.plot.bar(rot=0, color={"Month Temperature (°C)":"red","Month Average (°C)":"darkorange" })
ax.set_facecolor("lightgray")
```

```python
#Warmest month, birth rates 9 months later
Births = [7863]
Averages = [7902.000000]
index = ['Birth Rates - October 2017']
df = pd.DataFrame({'Month Birth Rate': Births, 
'Month Average': Averages}, index=index)
ax = df.plot.bar(rot=0, color={"Month Birth Rate":"green","Month Average":"darkorange"})
ax.set_facecolor("lightgray")
#findings: unaffected
```

```python
#Coldest month, birth rates 9 months later
Births = [7323]
Averages = [7553.700000]
index = ['Birth Rates - April 1999']
df = pd.DataFrame({'Month Birth Rate': Births, 
'Month Average': Averages}, index=index)
ax = df.plot.bar(rot=0, color={"Month Birth Rate":"green","Month Average":"darkorange"})
ax.set_facecolor("lightgray")
#findings: Decrease in Birth rates
```

```python
#Wettest month, driest month against average
Rainfall = [630.6, 16.2]
Averages = [137.780000, 79.693333]
index = ['Wettest Month - Feb 1990', 'Driest Month - August 1995']
df = pd.DataFrame({'Month Rainfall (mm)':Rainfall, 
'Month Average (mm)': Averages}, index=index)
ax = df.plot.bar(rot=0, color={"Month Rainfall (mm)":"blue","Month Average (mm)":"darkorange" })
ax.set_facecolor("lightgray")
```

```python
#Driest month, birth rates 9 months later
Births = [7801.200000]
Averages = [7476]
index = ['Birth Rates - May 1996']
df = pd.DataFrame({'Month Birth Rate': Births, 
'Month Average': Averages}, index=index)
ax = df.plot.bar(rot=0, color={"Month Birth Rate":"green","Month Average":"darkorange"})
ax.set_facecolor("lightgray")
#findings: Increase in birth rates
```

```python
#Rainfall and birth rates change over time (1990-2019)
x = (1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,20004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019)
y1 = Temp_data.mean(axis=1)
y2 = Birth_data.mean(axis=1)

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(y1, 'r-')
ax2.plot(y2, 'g-')

ax1.set_xlabel('Year')
ax1.set_ylabel('Temperature (°C)', color='r')
ax2.set_ylabel('Birth Rates', color='g')

plt.show()
#Increase in Birth rates as decrease in Temperature 
```
